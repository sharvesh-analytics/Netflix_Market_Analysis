## Outlier Detection

```python
Q1 = df["IMDb_Rating"].quantile(0.25)

Q3 = df["IMDb_Rating"].quantile(0.75)

IQR = Q3-Q1

lower = Q1-1.5*IQR

upper = Q3+1.5*IQR

df[
(df["IMDb_Rating"]<lower)
|
(df["IMDb_Rating"]>upper)
]
```
